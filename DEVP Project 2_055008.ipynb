```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to calculate and visualize statistics
def calculate_and_visualize(column_name):
    # Calculate statistics
    range_value = ncat[column_name].max() - ncat[column_name].min()
    std_dev = ncat[column_name].std()
    skewness = stats.skew(ncat[column_name])
    kurtosis = stats.kurtosis(ncat[column_name])
    correlation = ncat[column_name].corr(ncat['Quantity'])  # Assuming 'Quantity' is another relevant column
    coeff_var = std_dev / ncat[column_name].mean()
    confidence_interval = stats.t.interval(0.95, len(ncat[column_name]) - 1, loc=ncat[column_name].mean(), scale=std_dev / (len(ncat[column_name]) ** 0.5))

    # Print the results
    print(f"Column: {column_name}")
    print(f"Range: {range_value}")
    print(f"Standard Deviation: {std_dev}")
    print(f"Skewness: {skewness}")
    print(f"Kurtosis: {kurtosis}")
    print(f"Correlation with Quantity: {correlation}")
    print(f"Coefficient of Variation: {coeff_var}")
    print(f"Confidence Interval: {confidence_interval}")

    # Visualize
    fig, axes = plt.subplots(3, 2, figsize=(12, 8))
    fig.suptitle(f"Statistical Analysis for {column_name}")

    # Histogram
    axes[0, 0].hist(ncat[column_name], bins=30)
    axes[0, 0].set_title("Histogram")

    # QQ Plot
    stats.probplot(ncat[column_name], plot=axes[0, 1])
    axes[0, 1].set_title("QQ Plot")

    # Box Plot
    axes[1, 0].boxplot(ncat[column_name])
    axes[1, 0].set_title("Box Plot")

    # Scatter Plot (if there's a correlation)
    if correlation != 0:
        axes[1, 1].scatter(ncat['Quantity'], ncat[column_name])
        axes[1, 1].set_title(f"Scatter Plot (Correlation: {correlation:.2f})")
    else:
        axes[1, 1].text(0.5, 0.5, "No significant correlation", ha="center", va="center")
        axes[1, 1].axis('off')

    # Confidence Interval
    axes[2, 0].hlines(ncat[column_name].mean(), xmin=0, xmax=1, colors='blue', linestyles='dashed')
    axes[2, 0].hlines(confidence_interval[0], xmin=0, xmax=1, colors='green', linestyles='dashed')
    axes[2, 0].hlines(confidence_interval[1], xmin=0, xmax=1, colors='green', linestyles='dashed')
    axes[2, 0].set_title("Confidence Interval")

    plt.tight_layout()
    plt.show()

# Apply the function to each column
columns = ['Quantity', 'Value', 'Weight']
for column in columns:
    calculate_and_visualize(column)
```

```python
def analyze_categorical_column(column_name):
    """
    Analyzes a categorical column and prints the results.

    Args:
        column_name: The name of the column to analyze.
    """

    # Count, Frequency, Proportion
    value_counts = cat[column_name].value_counts()
    total_count = len(cat)

    print(f"\n**Column: {column_name}**")
    print("Count:\n", value_counts)
    print("\nFrequency:\n", value_counts / total_count)
    print("\nProportion:\n", value_counts / total_count)

    # Minimum, Maximum
    print("\nMinimum Value:", value_counts.min())
    print("\nMaximum Value:", value_counts.max())

    # Mode (Option 1: Using scipy.stats.mode - Recommended for SciPy < 1.11.0)
    # mode_result = scipy.stats.mode(cat[column_name])
    # print("Mode:", mode_result.mode[0])

    # Mode (Option 2: Using np.unique - Recommended for SciPy >= 1.11.0 or alternative)
    mode_counts = np.unique(cat[column_name], return_counts=True)[1].argmax()
    mode_value = np.unique(cat[column_name], return_counts=True)[0][mode_counts]
    print("\nMode:", mode_value)

    # Rank
    print("\nRank:", value_counts.rank(ascending=False))

    # Visualization
    if len(value_counts) <= 10:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=column_name, data=cat)
        plt.title(f"Distribution of {column_name}")
        plt.show()
    else:
        top_10 = value_counts.nlargest(10)  # Get top 10 categories
        others = value_counts[10:].sum()  # Sum of all other categories
        pie_labels = top_10.index.tolist() + ['Other']
        pie_values = top_10.tolist() + [others]

        # Calculate percentage for each category
        pie_percentages = [(value / total_count) * 100 for value in pie_values]

        # Create labels with percentages for the legend
        legend_labels = [f"{label}: {value:.1f}%" for label, value in zip(pie_labels, pie_percentages)]

        # Explode to create separation between the top 10 slices
        explode = [0.05] * len(pie_values)  # Small separation for each slice

        # Define a custom color palette and ensure "Others" has a unique color
        colors = sns.color_palette('Set3', len(pie_values) - 1) + ['#808080']  # Grey color for 'Other'

        plt.figure(figsize=(10, 6))
        wedges, texts = plt.pie(pie_values, explode=explode, colors=colors, startangle=140)

        # Add legend with category names and their respective percentages
        plt.legend(wedges, legend_labels, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

        plt.title(f"\nTop 10 Distribution of {column_name} (Others grouped)")
        plt.show()

categorical_columns = ['Payment_Terms', 'Country', 'Product', 'Import_Export', 'Category', 'Customs_Code', 'Shipping_Method', 'Supplier', 'Customer']

for column in categorical_columns:
    analyze_categorical_column(column)
```

```python
#Measures of Central Tendency

# Minimum and Maximum
min_values = ncat.min()
max_values = ncat.max()

# Mean and Median
mean_values = ncat.mean()
median_values = ncat.median()

# Mode
mode_values = ncat.mode()

# Percentile (for example, 25th, 50th, 75th)
percentiles = ncat.quantile([0.25, 0.5, 0.75])

print("Min:") 
print(min_values)
print("\nMax:") 
print(max_values)
print("\nMean:") 
print(mean_values)
print("\nMedian:") 
print(median_values)
print("\nMode:") 
print(mode_values)
print("\nPercentiles:\n") 
print(percentiles)
```

```python
#Finding Measures of Dispersion

# Range (Max - Min)
range_values = ncat.max() - ncat.min()

# Standard Deviation
std_dev = ncat.std()

# Skewness (measure of asymmetry)
skewness = ncat.skew()

# Kurtosis (measure of tails/heaviness)
kurtosis = ncat.kurt()

# Correlation matrix (Pearson correlation by default)
correlation_matrix = ncat.corr()

print("Range:") 
print(range_values)
print("\nStandard Deviation:") 
print(std_dev)
print("\nSkewness:") 
print(skewness)
print("\nKurtosis:")
print(kurtosis)
print("\nCorrelation:") 
print(correlation_matrix)
```

```python
#Composite Measure

# Coefficient of Variation (CV = std_dev / mean)
cv = std_dev / mean_values

# Confidence Interval (assuming a 95% confidence interval)
conf_interval = {}
for column in ncat.columns:
    mean = ncat[column].mean()
    sem = stats.sem(ncat[column])  # Standard error of the mean
    conf_interval[column] = stats.t.interval(0.95, len(ncat[column]) - 1, loc=mean, scale=sem)

print("Coefficient of Variation:")
print(cv)
print("\nConfidence Interval:")
print(conf_interval)
```

```python
#Test of Mean

# t-test (Comparing column Weight with a population mean of 5)
t_stat, p_val = stats.ttest_1samp(ncat['Weight'], 50)

# ANOVA (Comparing multiple groups)
f_stat, p_val_anova = stats.f_oneway(ncat['Weight'], ncat['Value'])

print("t-test p-value:")
print(p_val)

print("\nANOVA p-value:")
print(p_val_anova)
```

```python
#Test of Variance

# F-test (between columns A and B)
f_stat, p_val_f = stats.levene(ncat['Quantity'], ncat['Weight'])

# Levene's test (for equal variances)
levene_stat, levene_p = stats.levene(ncat['Quantity'], ncat['Value'])

# Bartlett’s test (for equal variances)
bartlett_stat, bartlett_p = stats.bartlett(ncat['Quantity'], ncat['Value'])

print("F-test p-value:")
print(p_val_f)
print("\nLevene test p-value:")
print(levene_p)
print("\nBartlett test p-value:")
bartlett_p
```

```python
#Test of Proportion

#Example: 10 successes out of 50 trials
successes = 10
n = 50
p0 = 0.5  # Hypothesized population proportion
z_stat, p_val_z = proportions_ztest(successes, n, value=p0)

# Chi-square test (for independence)
chi2_stat, p_val_chi2, _, _ = stats.chi2_contingency(pd.crosstab(ncat['Quantity'], ncat['Value']))

print("z-test p-value:", p_val_z)
print("\nChi-square p-value:\n") 
print(p_val_chi2)
```

```python
# Testing correlation significance
r, p_val_corr = stats.pearsonr(ncat['Value'], ncat['Quantity'])

print("Correlation p-value:", p_val_corr)
```

```python
# Shapiro-Wilk Test
shapiro_stat, p_val_shapiro = stats.shapiro(ncat['Quantity'])

# Kolmogorov-Smirnov Test
ks_stat, p_val_ks = stats.kstest(ncat['Weight'], 'norm', args=(ncat['Quantity'].mean(), ncat['Value'].std()))

# Anderson-Darling Test
anderson_stat = stats.anderson(ncat['Quantity'], dist='norm')

# Jarque-Bera Test
jb_stat, p_val_jb = stats.jarque_bera(ncat['Value'])

print("Shapiro-Wilk p-value:", p_val_shapiro)
print("\nKolmogorov-Smirnov p-value:", p_val_ks)
print("\nAnderson-Darling statistic:", anderson_stat)
print("\nJarque-Bera p-value:", p_val_jb)
```

```python
# Mann-Whitney U Test
mannwhitney_stat, p_val_mw = stats.mannwhitneyu(ncat['Value'], ncat['Quantity'])

# Wilcoxon Signed-Rank Test
wilcoxon_stat, p_val_wilcoxon = stats.wilcoxon(ncat['Weight'] - ncat['Quantity'])

# Kruskal-Wallis Test (ANOVA alternative)
kruskal_stat, p_val_kruskal = stats.kruskal(ncat['Value'], ncat['Weight'])

# Friedman Test (for repeated measures)
friedman_stat, p_val_friedman = stats.friedmanchisquare(ncat['Value'], ncat['Weight'], ncat['Value'] + ncat['Weight'])

print("Mann-Whitney p-value:", p_val_mw)
print("Wilcoxon p-value:", p_val_wilcoxon)
print("Kruskal-Wallis p-value:", p_val_kruskal)
print("Friedman p-value:", p_val_friedman)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# List of categorical columns
categorical_columns = ['Payment Terms', 'Country', 'Product', 'Import_Export', 
                       'Category', 'Customs_Code', 'Shipping_Method', 'Supplier', 'Customer']

# Function to calculate statistics
def calculate_stats(df, column):
    print(f"\nStatistics for '{column}':")
    
    # Count
    count = df[column].count()
    print(f"Count: {count}")
    
    # Frequency (value counts)
    freq = df[column].value_counts()
    print(f"Frequency:\n{freq}")
    
    # Proportion (percentage of each category)
    proportion = df[column].value_counts(normalize=True) * 100
    print(f"Proportion (percentage):\n{proportion}")
    
    # Minimum
    min_value = df[column].min()
    print(f"Minimum: {min_value}")
    
    # Maximum
    max_value = df[column].max()
    print(f"Maximum: {max_value}")
    
    # Mode
    mode_value = df[column].mode()[0]
    print(f"Mode: {mode_value}")
    
    # Rank (based on frequency)
    rank = df[column].value_counts().rank(ascending=False)
    print(f"Rank:\n{rank}")
    
    return freq, proportion

# Function to plot visualization
def plot_visualization(df, column, freq):
    plt.figure(figsize=(8, 6))
    unique_vals = df[column].nunique()
    
    if unique_vals > 10:
        # Bar plot for columns with more than 10 unique values
        sns.barplot(x=freq.values, y=freq.index, palette='viridis')
        plt.title(f'Bar Plot of {column}', fontsize=14)
        plt.xlabel('Frequency', fontsize=12)
        plt.ylabel(column, fontsize=12)
        plt.grid(True, axis='x', linestyle='--', alpha=0.7)
    else:
        # Pie chart for columns with 10 or fewer unique values
        freq.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis', unique_vals))
        plt.title(f'Pie Chart of {column}', fontsize=14)
        plt.ylabel('')
    
    plt.tight_layout()
    plt.show()

# Function to calculate correlation for categorical variables (Cramér's V)
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

def calculate_correlation(df, column):
    for other_col in categorical_columns:
        if other_col != column:
            # Create confusion matrix
            confusion_matrix = pd.crosstab(df[column], df[other_col])
            
            # Only calculate correlation if both columns have more than one unique category
            if confusion_matrix.shape[0] > 1 and confusion_matrix.shape[1] > 1:
                corr = cramers_v(confusion_matrix)
                print(f"Cramér's V Correlation between '{column}' and '{other_col}': {corr:.4f}")
            else:
                print(f"Skipping correlation calculation for '{column}' and '{other_col}' (only one category).")

# Main loop to run the analysis on each column
for col in categorical_columns:
    try:
        freq, proportion = calculate_stats(cat, col)  # Calculate statistics
        plot_visualization(cat, col, freq)  # Plot visualization
        calculate_correlation(cat, col)  # Calculate correlation
    except Exception as e:
        print(f"Error processing column '{col}': {e}")


categorical_columns = ['Payment_Terms', 'Country', 'Product', 'Import_Export', 'Category', 'Customs_Code', 'Shipping_Method', 'Supplier', 'Customer']

for column in categorical_columns:
    analyze_categorical_column(column)
```
